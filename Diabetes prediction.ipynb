```python
# importing the requirements
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# import the data
df = pd.read_csv("/content/diabetes.csv")
df.head()
```

```python
# check for null values
df.isnull().sum()
```

```python
# dataset info
df.info()
```

```python
#splitting x and y
x = df.drop("Outcome", axis=1)
y = df["Outcome"]
x
```

```python
y
```

```python
# splitting the train and test data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = Sequential() #creates a linear stack
model.add(Dense(8,activation = "relu", input_shape=(8,))) #adds fully connected first layer of neurons with 8units of neurons
model.add(Dense(8,activation = "relu")) #adds a second layer & keras  takes the input value automatically
model.add(Dense(1,activation = "sigmoid")) # adds final layer with one neuron i.e 0 or 1
```

```python
# this line compiles the model and calculates the loss and implement optimizer
model.compile(loss = "binary_crossentropy", optimizer = "sgd",metrics = ["accuracy"])
# run 5 iterations
model.fit(x_train, y_train, epochs = 5, batch_size = 1, verbose = 1)
```

```python
# this line evaluate score of the model
score = model.evaluate(x_test, y_test, verbose = 1)
print(score)
```
